```python
from webull import webull
import numpy as np
import pandas as pd
import datetime
import xlsxwriter
import yfinance as yf 
import requests

```

```python
wb = webull()
```

```python
stock = ['SPCB', 'HOFV', 'DSS', 'KOPN', 'ZKIN', 'CLVS', 'TKAT', 'XSPA', 'NAKD', 'UAVS', 'SNES', 'OCG','JFIN', 'DGLY', 'JE', 'AAL', 'LAZR', 'BOXL', 'VISL', 'WIMI', 'AKER', 'ORMP', 'DM', 'NNDM', 'SOLO', 'MVIS', 'SOL', 'IZEA', 'OEG', 'SUNW', 'SPWR', 'GNUS', 'UNFI', 'ASXC', 'YALA', 'GEVO', 'COTY', 'VALE', 'CLNE', 'IDEX', 'AMRS', 'AMKR', 'FSR', 'BNGO', 'BLNK', 'CODX', 'TME', 'SHIP', 'DOYU', 'HIMX', 'FINV', 'EYES', 'OCGN', 'NNVC', 'ETH', 'MOXC', 'WKEY', 'LIZI', 'TAOP', 'SEEL', 'CETX', 'SNCA', 'EVOL', 'IMTE', 'SOS', 'HEPA', 'AESE', 'SNDL', 'LOTZ', 'ENVB', 'LODE', 'FSI', 'PIXY', 'CMO', 'GORO', 'SWN', 'VUZI', 'CCL', 'PLTR']
```

```python

#####SET CORRECT YF DAYS
maxup = []
maxdown_matrix = [] 
min_before_max_matrix = []
workbook  = xlsxwriter.Workbook('/Users/jpjansen/Documents/3_21WBPerformance_V2.xlsx')
worksheet = workbook.add_worksheet()
worksheet.write(0,0,'Stock Symbol')
worksheet.write(0,1,'Max Up')
worksheet.write(0,2,'Max Down')
worksheet.write(0,3,'Min before max')
#days = input("Type in period of days for min before max in format (#)d")
for i in range(len(stock)):
    print(stock[i])
    try:
        weekly_data = wb.get_bars(stock[i], interval= 'w1',extendTrading=0, count =3)
        
        this_week = weekly_data.loc['2021-03-26 00:00:00-04:00']
        maxup = 100*(this_week['high']-this_week['open'])/this_week['open']
        maxdown = 100*(this_week['open']-this_week['low'])/this_week['open']
        worksheet.write(i+1,0,stock[i])
        worksheet.write(i+1,1,maxup)
        worksheet.write(i+1,2,maxdown)
        
    except KeyError:
        worksheet.write(i+1,0,stock[i])
        worksheet.write(i+1,1,'Key Error')
        worksheet.write(i+1,2,'Key Error')
    except ValueError:
        worksheet.write(i+1,0,stock[i])
        worksheet.write(i+1,1,'Key Error')
        worksheet.write(i+1,2,'Key Error')
for i in range(len(stock)):
    try:
        data = yf.download(tickers=stock[i], period="5d", interval="1m")
        iterate = data.to_numpy()
        min_before_max = np.amin(iterate[0:np.argmax(iterate[:,1]),2])
        open_price = iterate[0,0]
        min_before_max_pct = ((min_before_max-open_price)/open_price)
        min_before_max_matrix.append((min_before_max-open_price)/open_price)
        worksheet.write(i+1,3,min_before_max_pct)
    except KeyError:
        worksheet.write(i+1,3,'Key error')     
    except ValueError:
        worksheet.write(i+1,3,'Key error')
workbook.close()
```
